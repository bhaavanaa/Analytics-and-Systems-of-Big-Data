***PROBLEM - 3***

**Problem statement** - Binarize the attribute “Year”. Set the threshold above 2016 and print it without truncation. (Use AVOCADO dataset)

**Logic used** - If the year attribute of the dataframe is less than 2016, then the Binarize_year attribute will have 0, else it will have 1.

**Packages used** - Pandas - to perform operations on the dataset which is loaded as a dataframe.

```
import pandas as pd

avocado = pd.read_csv("Avocado_Dataset.csv")

binarize = []
for year in avocado['year']:
  if year > 2016 : binarize.append(1)
  else : binarize.append(0)

avocado['Binarize_year'] = binarize
avocado
```

***PROBLEM - 4*** 

**Problem statement** - Transform all categorical attributes in the dataset AVOCADO using Integer Encoding.

**Logic used** - The categorical attributes in the dataset are 'type' and 'region'. So every category is given an integer and a new attributes are added with their respective integer encoding obtained for the values present in the row. 

**Packages used** - Pandas - to perform operations on the dataset which is loaded as a dataframe.

```
import pandas as pd

avocado = pd.read_csv("Avocado_Dataset.csv")

type_categories = list(set(avocado['type']))
region_categories = list(set(avocado['region']))

print("Encoding for type -")
for i in range(len(type_categories)):
  print(type_categories[i], " : ", i)

print("\nEncoding for region -")  
for i in range(len(region_categories)):  
  print(region_categories[i], " : ", i)

type_encoding = []
region_encoding = []

for i in range(avocado.shape[0]):
  type_encoding.append(type_categories.index(avocado['type'][i]))
  region_encoding.append(region_categories.index(avocado['region'][i]))

avocado['Type_encoding'] = type_encoding
avocado['Region_encoding'] = region_encoding

avocado
```

***PROBLEM - 5***

**Problem statement** - Transform the attribute = “Region” in the given dataset AVOCADO using One-Hot Encoding.

**Logic used** - Obtain the list of unique regions and get their one-hot encodings using LabelBinarizer and add the 'One_hot_encoding_region' attribute to the dataframe.

**Packages used** - 
*   Pandas - to perform operations on the dataset which is loaded as a dataframe.
*   Sklearn - to obtain the one hot encodings of the regions.



```
import pandas as pd
from sklearn.preprocessing import LabelBinarizer 

avocado = pd.read_csv("Avocado_Dataset.csv")

regions = list(set(avocado['region']))

label_binarizer = LabelBinarizer()
label_binarizer.fit(regions)
one_hot_encoding = label_binarizer.transform(regions)

for i in range(len(regions)):
  print(regions[i], " : ", ''.join(map(str, one_hot_encoding[i])))

region_onehot_encoding = []
for region in avocado['region']:
  region_onehot_encoding.append(''.join(map(str, one_hot_encoding[regions.index(region)])))

avocado['One_hot_encoding_region'] = region_onehot_encoding
avocado
```

***PROBLEM - 6***

**Problem statement** - Ignore the tuples that hold missing values and print the subset of data from AVOCADO dataset.

**Logic used** - Store the indices of the rows which contain null values and remove all such rows using df.drop() function.

**Packages used** - Pandas - to perform operations on the dataset which is loaded as a dataframe.

```
import pandas as pd

def check(value):
  if(str(value).startswith('n') or str(value).startswith('N') or str(value) == ''):
    return False
  else:
    return True

avocado = pd.read_csv("Avocado_Dataset.csv")

print("The number of rows in the original dataset : ", avocado.shape[0])

indices = []
for i in range(avocado.shape[0]):
  if(check(avocado['AveragePrice'][i]) == False):
    indices.append(i)

rows = avocado.index[indices]
avocado.drop(rows, inplace=True)

print("The number of rows in the modified dataset : ", avocado.shape[0])

avocado
```

***PROBLEM - 7***

**Problem statement** - Drop the attribute that has high nullity as it facilitates efficient prediction. (Use AVOCADO dataset)

**Logic used** - Find the columns wich contain NaN values and obtain the count of the number of NaN values in each column and drop the column which contains the max number of NaN values.

**Packages used** - Pandas - to perform operations on the dataset which is loaded as a dataframe.

```
import pandas as pd

avocado = pd.read_csv("Avocado_Dataset.csv")

nan_columns = avocado.columns[avocado.isna().any()].tolist()

count = []
for column in nan_columns:
  count.append(avocado[column].isna().sum())

avocado.drop([nan_columns[count.index(max(count))]], axis = 1, inplace=True) 

print("The column with max number of NaN values : ", nan_columns[count.index(max(count))])

avocado
```

**Problem Statement** - Test drive the implementation support in your platform of choice for data preprocessing phases such as cleaning, selection, transformation, integration in addition to the earlier exercises.

**Logic used** - Cleaning is done by replacing null values with mean and median, selection is done using entropy and information gain as done in Problem 9, and transformation is done using min-max and z-score.

**Packages used** - 
* Pandas - to perform operations on the dataset which is loaded as a dataframe.
* Statistics - to compute mean and median.

```
import pandas as pd
import statistics

trail = pd.read_csv("Trail.csv")

z=[]
for i in range (trail.shape[0]):
  x = trail['AveragePrice'][i]
  if(str(x).startswith('n') or str(x).startswith('N') or str(x) == "" ):
    continue
  else:
    z.append(float(trail['AveragePrice'][i]))

m1 = statistics.mean(z)
m2 = statistics.median(z)

# Cleaning type 1 - using mean for null values
for i in range (trail.shape[0]):
  x = trail['AveragePrice'][i]
  if(str(x).startswith('n') or str(x).startswith('N') or str(x) == "" ):
    trail['AveragePrice'] = m1

# Cleaning type 2 - using median for null values
for i in range (trail.shape[0]):
  x = trail['AveragePrice'][i]
  if(str(x).startswith('n') or str(x).startswith('N') or str(x) == "" ):
    trail['AveragePrice'] = m2

### Selection is done in PROBLEM - 9 using the Entropy and Information gain
```
